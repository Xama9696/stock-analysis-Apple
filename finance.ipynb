```python
import pandas as pd

df = pd.read_csv("HistoricalData_1774955312443.csv")
df.head()
```

```python
df.columns
```

```python
df.rename(columns={
    "Date": "Date",
    "Close/Last": "Close",
    "Volume": "Volume",
    "Open": "Open",
    "High": "High",
    "Low": "Low"
}, inplace=True)
```

```python
df["Close"] = df["Close"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["returns"] = df["Close"].pct_change()

df["MA_7"] = df["Close"].rolling(7).mean()
df["MA_30"] = df["Close"].rolling(30).mean()

df["volatility"] = df["returns"].rolling(7).std()
```

```python
cols = ["Open", "High", "Low"]

for col in cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)
```

```python
df["Close"] = df["Close"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.to_csv("finance_dashboard.csv", index=False)
```
